# REGRESI LINIER


Regresi Linear adalah Metode Statistik yang berfungsi untuk menguji sejauh mana hubungan sebab akibat antara Variabel Faktor Penyebab (X) terhadap Variabel Akibatnya. Tujuan regresi linier untuk mendapatkan pola hubungan secara matematis dari variabel X dan variabel Y, dan untuk mengetahui besarnya perubahan variabel X terhadap variabel Y, serta untuk memprediksi variabel Y jika nilai variabel X diketahui.

```r
#Model Regresi Linear Sederhana dari 2 variabel dengan korelasi terkuat yang didapatkan dari matriks korelasi
ggplot(data,aes(ApplicantIncome, LoanAmount)) +
  geom_point() +
  geom_smooth(method='lm', se=FALSE, color='turquoise4') +
  theme_minimal() +
  labs(x='ApplicantIncome', y='LoanAmount', title='Linear Regression Plot') +
  theme(plot.title = element_text(hjust=0.5, size=20, face='bold')) 
```

### **Hasil Analisis :**
Grafik/Diagram di atas menunjukkan hubungan yang meningkat secara linier antara variabel ApplicantIncome dan LoanAmount. Ini adalah hal yang baik, karena salah satu asumsi penting dari regresi linier adalah bahwa hubungan antara variabel hasil dan prediktor adalah linier dan aditif atau konsisten naik. Meskipun masih banyak titik error yang menjauhi dari garis linier/prediksinya
